# Corso di Probabilit√† e Statistica - Plot
In questo capitolo introduciamo alcuni metodi per fare dei semplici grafici in python

```python
# importare packages
import numpy as np #operazioni numeriche
import matplotlib.pyplot as plt #grafici
```

### Plottare una funzione

#### Generazione e salvataggio dei dati

```python
# generare i dati (x = cos(t)  con t in [a,b], dt=0.1)
a=0
b=2*np.pi # np.pi = pi greco
dt=0.1

t = np.arange(a,b,dt)
x = np.cos(t)
```

```python
plt.plot(t,x)
plt.xlabel('t')
plt.ylabel('x(t)')

plt.show()
```

```python
# salvare i dati in un file di testo
out_file = '../data/cosine.txt' #gli ho detto in pratica dalla cartella in cui sono alla cartella data
np.savetxt(out_file, np.column_stack([t,x]) )
```

#### Plot di dati da file

```python
# leggere i dati in una nuova matrice
inData = np.loadtxt(out_file) #NB i dati che sto caricando sono salvati in formato txt

t2 = inData[:,0] # Nota Python comincia da "0"!
x2 = inData[:,1]
#leggo tutte le righe fissando la colonna da cui li voglio leggere
```

```python
#fare un grafico dei dati
plt.plot(t2, x2)
plt.xlabel('t')
plt.ylabel('x(t)')

plt.show()
```

```python
# alternativa
plt.plot(inData[:,0],inData[:,1])
#gli dico che deve leggere la variabile indipendente dalla prima colonna e quella dipendente dalla seconda

plt.xlabel('t')
plt.ylabel('x(t)')
plt.show()
```

```python
#Attenzione se plottiamo direttamente la matrice otteniamo questo:
plt.plot(inData)
```

Cio√® in realt√† non √® nemmeno cos√¨ terribile, mi fa due plot per le due colonne che vede assumendo che vadano plottati entrambi rispetto allo stesso set di variabili indipendenti.
Quelle che prima erano variabili indipendenti diventano una retta, quello che prima era un coseno rimane un coseno ma deformato per via della scelta dei valori degli assi che fa lui.

**Aggiungere variabili sugli assi**

```python
plt.xlabel("Tempo")
plt.ylabel("Posizione")

plt.plot(t, np.cos(t))
plt.show()
```

**Rendere il grafico leggibile**

```python
plt.rcParams.update({'font.size': 18})
#ho fissato la dimensione del testo, in questo modo scrive solo alcune delle unit√† di misura sugli assi, quelle che ci stanno col nuovo font, per il resto ovviamente non cambia nulla
plt.xlabel("Tempo")
plt.ylabel("Posizione")
plt.plot(t, np.cos(t))
plt.show()
```

**Aggiungere il titolo**

```python
plt.rcParams.update({'font.size': 12})
plt.xlabel("Tempo")
plt.ylabel("Posizione")
plt.title("Oscillatore armonico")
plt.plot(t, np.cos(t))
plt.show()
```

**Aggiungere una legenda**

```python
plt.rcParams.update({'font.size': 12})
plt.xlabel("Tempo")
plt.ylabel("Posizione")
plt.title("Oscillatore armonico")
plt.plot(t, np.cos(t))
plt.legend("x")

plt.show()
```

**Due curve**

```python
plt.rcParams.update({'font.size': 12})
plt.xlabel("Tempo")
plt.ylabel("Posizione")
plt.title("Oscillatore armonico")
plt.plot(t, np.cos(t), label="x1")
plt.plot(t, np.cos(2*t), label="x2") #oscillatore con oscillazione diversa
plt.legend()
plt.show()
```

**Spostare la legenda**

```python
plt.rcParams.update({'font.size': 12})
plt.xlabel("Tempo")
plt.ylabel("Posizione")
plt.title("Oscillatore armonico")
plt.plot(t, np.cos(t), label="x1")
plt.plot(t, np.cos(2*t), label="x2")
plt.legend(loc="upper left")
# oppure "upper right", "center", etc

plt.show()
```

### Esercizio

Considerare le due funzioni: 
f(x) = sin(x)*exp(-x) e
g(x) = cos(x)*exp(-x)

- Generare i dati con le due funzioni in [0,5] e salvarli in due files 
- Fare un grafico che mostri le due funzioni sempre in [0,5]



```python
#definisco le funzioni e le variabili
x = np.linspace(0,5,500)
f = np.exp(-x)*np.sin(x)
g = np.exp(-x)*np.cos(x)

#ipmosto il grafico
plt.rcParams.update({'font.size': 12})

plt.plot(x,f, label = "f(x)", color = "lightblue")
plt.plot(x,g, label = "g(x)", color = "blue")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Oscillatore armonico smorzato')
plt.legend()
plt.show()

#salvo il grafico
name = "Oscillatore armonico smorzato.png"
plt.savefig(name)
```
