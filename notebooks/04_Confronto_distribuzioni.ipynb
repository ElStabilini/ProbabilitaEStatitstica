# Corso di Probabilit√† e Statistica - Confrontare set di dati diversi

In questo capitolo introduciamo alcuni metodi per visualizzare nello stesso grafico diversi set di dati



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats # Statistica
import seaborn as sns # Grafica avanzata
```

```python
# usiamo caratteri leggibili!
plt.rcParams.update({'font.size': 14})
```

```python
# Generare 3 set di N numeri random Gaussiani 
N=100

#in pratica sto generando 3 set di numeri distribuiti ciascuno secondo una gaussiana ma gaussiane diverse

sigma1=1. #sigma della distribuzione
m1=1.5 #media della distribuzione
x1 = sigma1*np.random.randn(N)+m1

sigma2=2.
m2=2.
x2 = sigma1*np.random.randn(N)+m2

sigma3=3.
m3=1.
x3 = sigma3*np.random.randn(N)+m3
```

**Nota:** Tutti i tipi di grafici che sono qui riportati possono essere rappresentati anche in orizzontale

**Istogrammi**

```python
plt.hist(x1,bins=20,density=True,label="set1")
plt.hist(x2,bins=20,density=True, label="set2")
plt.hist(x3,bins=20,density=True, label="set3")

plt.legend()
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Histograms')
plt.show()
```

Di seguito illustriamo metodi per visualizzare pi√π semplicemente tante distribuzioni insieme

```python
# matrice complessiva (array) utile se x1,x2,x3 hanno la stessa dimensione
data=[x1,x2,x3]
#organizzo i 3 set di dati all'interno di un array
```

**Boxplot**

https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
# The error bars indicate 1.5* the inter-quartile-range (IQR), and the box consists of the
# first, second (middle) and third quartile
sns.boxplot(data=data, width=0.4, palette='Blues')
plt.title('Boxplot')
plt.xlabel('Index')
plt.ylabel('X')
plt.draw()
```

##### Quartili

In statistica, un *quartile* √® un tipo di *quantile* che divide il numero di dati in quattro parti (o quarti) pi√π o meno della stessa dimnesione (lo stesso numero di dati in ciascuna porzione in cui li ho suddivisi). Per poter utilizzare la suddivisione in quartili i dati devono essere ordinati dal pi√π piccolo al pi√π grande; in questo modo i quartili sono una forma di statistica ordinata. I tre quartili principali sono i seguenti: 

* Il primo quartile (Q1) √® definito come il **valore medio tra il numero pi√π piccolo** (minimo) della distribuzione e la  **mediana** del data-set. √® anche conosciuto come *quartile inferiore* o *25-esimo quartile empirico* dato che il 25% dei dati si trova al di sotto di questo punto.
* Il secondo quartile (Q2) √® la mediana del set di dati; quindi il 50% dei dati si trova al di sotto di questo punto.
* Il terzo quartile (Q3) √® il valore medio tra la mediana e il valore massimo del data set. √® conosciuto anche come il quartile *superiore* o *75-esimo quartile empierico*, dato che il 75% dei dati si trova al di sotto di questo punto 

Insieme al massimo e al minimo dei dati i tre quartili (che sono quindi 3 punti fissati all'interno della distribuzione di dati) forniscono anche un "riassunto" in cinque numeri della distribuzione di dati.
Questo "riassunto" √® importante in statistica perch√® fornisce informazioni sia sul **centro** che sulla **distribuzione** dei dati. Conoscere il quartile inferiore e quello superiore fornisce informazioni su quanto sono *allargati* i dati e **se la distribuzione √® concentrata pi√π verso un estremo piuttosto che un altro**. Dato che i quartili dividono i dati in maniera "uniforme" la distranza tra i quartili non √® sempre la stessa (i.e., Q3-Q2 ‚â† Q2-Q1) ma √® conosciuta come *inter-quartile range* (IQR).  Mentre il massimo e il minimo mostrano comunque la distribuzione dei dati, il quartile superiore e inferiore forniscono informazioni pi√π dettagliate sulla posizione di alcuni punti specifici: la presenza di outliers all'interno dei dati e la differenza nella distribuzione tra il punto centrale della distribuzione e i dati pi√π lontani.

#### Grafico a box
I limiti del box sono rispettivamente il primmo, il secondo e il terzo quartile. La barra d'errore indica 1.5 $\cdot$ IQR, i baffi si estendono per indicare il resto della distribuzione

### Swarmplot

https://seaborn.pydata.org/generated/seaborn.swarmplot.html

Questo tipo di plot √® pi√π adatto alla rappresentazione di un altro tipo di dati, ad esempio posso suddividire uno stesso dataset (eg. - che si trova anche online - dimanati) in pi√π sotto-gruppi sulla base di una determinata caratteristica (eg. - dimensione dei diamanti) e vedere come una seconda caratteristica √® distribuita rispetto alla prima (eg. - costo dei diamanti).

Oppure suddivido i dati sulla base di una determinata categoria e poi li distribuisco in funzione di una seconda (eg: asse x metto diversi tipi di taglio del diamante e asse y il prezzo). La denist√† dei punti rappresenta quanti diamanti soddisfano le condizioni determinate dalla sovrapposizione di entrambe le categorie.

In questo caso √® assolutamente inutile utilizzare questo tipo di plot. La densit√† dei punti mostra, per ciascuna distribuzione - asse x - attorno a quale valore si concentrano i dati.

```python
sns.swarmplot(data=data, palette='Greens')
plt.title('Swarmplot')
plt.xlabel('Index')
plt.ylabel('X')
plt.draw()
```

### Violinplot

https://seaborn.pydata.org/generated/seaborn.violinplot.html

Ovviamente anche i plot a violino sono adatti a lavorare con diverse categorie di dati, ad esempio in questo caso stiamo guardando diverse distribuzioni di dati.

√® anche possibile effettuare degli split dei grafici a violino per, ad esmepio, due sottocategorie per la stessa categoria (eg. considero diversi tipi di fiori, una sottocategoria dei fiori pu√≤ essere maschi e femmine).

Il **punto bianco centrale** rapperesenta la mediana, la **barra grigia spessa** rappresenta il range interquartile (distanza Q1-Q3), in particolare la parte larga colorata fa capire come sia distribuita la probabilit√† (zona pi√π larga dove √® pi√π probabile trovare dati, ovviamente la parte colorata √® normalizzata - pi√π √® larga pi√π √® corta e viceversa.

La **linea grigia sottile** rappresenta il resto della distribuzione fatta eccezione per gli outlayers.

```python
sns.violinplot(data=data, palette='Oranges')
plt.title('Violinplot')
plt.xlabel('Index')
plt.ylabel('X')
plt.draw()
```

```python
sns.violinplot(data=data, inner="points", palette='Oranges')
plt.title('Violinplot')
plt.xlabel('Index')
plt.ylabel('X')
plt.draw()
```

### Stripplot

https://seaborn.pydata.org/generated/seaborn.stripplot.html

Generalmente l'utilit√† di questo tipo di plot √® quella di rivelare eventuali outlayers e gap all'interno della distribuzione di dati che si sta considerando.

```python
sns.stripplot(data=data, palette='viridis')
plt.title('Stripplot')
plt.xlabel('Index')
plt.ylabel('X')
plt.draw()
```

### Combinazioni di plot

```python
sns.boxplot(data=data, palette='viridis')
sns.stripplot(data=data, color="Turquoise")
plt.title('Stripplot & Boxplot')
plt.xlabel('Index')
plt.ylabel('X')
plt.draw()
```
